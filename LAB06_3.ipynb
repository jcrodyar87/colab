```python
# Define las columnas para el crosstab
columna_fila = "Region"  # Reemplaza con la columna para las filas
columna_columna = "Cards"  # Reemplaza con la columna para las columnas
valores = "Rating"  # Reemplaza con la columna para los valores

# Crea el crosstab
tabla_crosstab = pd.crosstab(
    df1[columna_fila],
    df1[columna_columna],
    values=df1[valores],
    #aggfunc=sum
    aggfunc=len
)

tabla_crosstab
```

```python
import numpy as np
# Define las columnas para el crosstab
columna_fila = "Age"  # Reemplaza con la columna para las filas
columna_columna = "Married"  # Reemplaza con la columna para las columnas
valores = "Income"  # Reemplaza con la columna para los valores

# Crea el crosstab
tabla_crosstab2 = pd.crosstab(
    df1[columna_fila],
    df1[columna_columna],
    values=df1[valores],
    #aggfunc=sum
    aggfunc=np.mean
)

tabla_crosstab2
```
