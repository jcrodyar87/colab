## **PLOTS DE DISTRIBUCION EN SEABORN**

## **distplot**

El distplot muestra la distribucion de un conjunto univariante de observaciones.

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(tips['total_bill']);
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10);
```

## **jointplot**

joinplot() permite basicamente emparejar dos distplots para datos bivariados, Con una eleccion de que paramtro **kind** va a comparar:
* scatter
* reg
* resid
* kde
* hex

```python
sns .jointplot(x='total_bill',y='tip',data=tips, kind='scatter');
```

```python
sns .jointplot(x='total_bill',y='tip',data=tips, kind='hex');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg');
```

## **PAIRPLOT**

Pairplot grafica de realaciones por pares en un dataframe completo(para las columnas numericas) y que soporta un argunento de el color de tono(hue)(para columnas cateogricas)

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm');
```
