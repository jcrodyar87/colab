# **CONTEXTO PROBLEMA**

Comprender cuál será el Burnout Rate para el empleado que trabaja en una organización en función de la **situación pandémica actual**, donde el trabajo desde casa es una bendición y una pesadilla. ¿Cómo se ve afectada la tasa de BURN de los empleados según las diversas condiciones proporcionadas?

**Conjunto de datos-Dataset**

Los siguientes son los atributos de datos y su descripción:

* **ID de empleado:** el ID único asignado por la empresa a cada empleado.Employee ID
* **Fecha de Ingreso:** La fecha en que el empleado se incorporó a la empresa.
Date of Joining

* **Género:** El sexo del empleado. Gender:
* **Tipo de empresa:** El tipo de empresa donde trabaja el empleado (Servicio/Producto). Company Type:
* **Configuración de la FMH disponible**:
 si las instalaciones de trabajo desde casa están disponibles para el empleado (Sí/No).WFH Setup Available
* **Designación**: La designación del empleado en su organización. En el rango: [0,0, 5,0], 0,0 es la designación más baja y 5,0 es la más alta. Designation
* **Asignación de Recursos:** La cantidad de recursos asignados al empleado para trabajar, que se interpretará como la cantidad de horas de trabajo. Dentro del rango: [1.0, 10.0] (más alto significa más recursos). Resource Allocation
* **Puntuación de fatiga mental:** cuánto cansado mentalmente está el empleado durante las horas de trabajo en el rango: [0,0, 10,0] donde 0,0 significa sin fatiga y 10,0 significa fatiga total. Mental Fatigue Score
* **Tasa de agotamiento:** el valor objetivo en los datos de cada empleado que indica la tasa de agotamiento durante las horas de trabajo en el rango: [0.0, 1.0] donde cuanto mayor es el valor, mayor es el agotamiento.  Burn Rate

# **BOXPLOT**

Un BoxPlot (o diagrama de caja, en traducción libre) muestra la distribución cuantitativa de los datos de una manera que facilita la comparación entre las variables, o a través de los niveles categóricos de las variables.

Esta caja ("box") muestra los cuartiles del conjunto de datos mientras que los "whiskers" muestran el resto de la distribución, excepto los puntos que se denominan valores atípicos.

cuartil.png

```python
import seaborn as sns
```

```python
sns.boxplot(x='Gender',y='Burn Rate',data=df)
```
