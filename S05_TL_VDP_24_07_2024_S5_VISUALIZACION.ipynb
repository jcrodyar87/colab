## **¿Por qué DASH?**

El uso de la biblioteca Plotly Python por sí mismo crea gráficos interactivos como archivos .html. Los usuarios pueden interactuar con estos gráficos (acercar, seleccionar, pasar el mouse sobre ellos), pero estos gráficos no se pueden conectar a fuentes de datos cambiantes. Debe volver a ejecutar el script .py y volver a generar el archivo .html para ver las actualizaciones. Tener que volver a ejecutar scripts .py no es realmente lo que queremos cuando pensamos en paneles. Queremos que se actualice automáticamente cada vez que cambien los datos. Aquí es donde necesitamos algo como Dash.

## **¿Que es DASH?**

Dash es una libreria de interfaz de usuario/framework de Python para crear aplicaciones web analíticas. Aquellos que usan Python para análisis de datos, exploración de datos, visualización, modelado, control de instrumentos e informes encontrarán un uso inmediato para Dash.

Está construido sobre Flask, Plotly.js, React y React Js. Le permite crear tableros utilizando Python puro. Dash es de código abierto y sus aplicaciones se ejecutan en el navegador web. Cada elemento estético de la aplicación es personalizable: el tamaño, el posicionamiento, los colores, las fuentes.

Dash es un framework de Python que se creó con el objetivo principal de hacer aplicaciones web, sin embargo el alcance de sus herramientas permite realizar y customizar tus dashboard o cuadros. El atractivo principal de esta herramienta es que ayuda a los científicos de datos a construir aplicaciones web analíticas e interactivas sin la necesidad de una instrucción en desarrollo web.

Dash basa su funcionamiento en tres tecnologías principales: la primera se encarga de suplir o proporcionar la funcionalidad que tendría el servidor web esta es “Flask; la segunda “React.js” toma la información de la interfaz página web y la renderiza, por último “Ploty.js” crea los gráficos usando los datos proporcionados. ## **¿Que herramientas me ofrece Dash?**

**Componentes HTML**

Dash funciona como un lugar para trabajar aplicaciones web que permite abstracción de Python en torno HTML, CSS y JavaScript, así en vez de tener que escribir código HTML basta con usar un motor de plantillas HTML con el módulo Dash HTML Components (dash.html), el cual es parte de Dash y del cual encontraras su código fuente en el repositorio GitHub de Dash.

**Componentes Dash Core**

Dash incorpora diferentes y variados componentes precargados para crear interfaces de usuarios interactivas, en este caso se usó el módulo (dash.dcc) que permite acceso entre diversas cosas a desplegables, listas de control y deslizadores.

**Callbacks Básicos**

Es importante mantener siempre la actualización de tu aplicación web, es por esto que Dash permite realizar funciones de callbacks básicos es decir funciones que son llamadas cada vez que se produce un cambio en algún componente de entrada para luego actualizar o realizar un cambio en algún elemento de salida.

**Añadiendo estilo CSS**

Dash permite añadir tus propios estilos CSS o JavaScript para esto se deberá crear una carpeta con el nombre “assets” en la raíz del directorio de la app y añadir en esta tus estilos predilectos de CSS o JavaScript ya que dash lee automáticamente todos los archivos que se encuentren en la mencionada carpeta.

## **Diseño de la aplicacion Dash**

Una aplicación Dash suele estar compuesta por 3 pilares:

1. **Dash Components:** componentes como el control deslizante, la casilla de verificación y el menú desplegable, para darle a su aplicación una capacidad interactiva.
2. **Plotly Graphs:** gráficos que permiten al usuario visualizar los datos
3. **The "Callback":** uno de los elementos más críticos. Conecta Dash Components y Plotly Graphs para crear una capacidad interactiva. Los componentes de Dash y los gráficos Plotly están dentro del diseño de la aplicación y describen cómo se verá la aplicación, y la a traves de callback se define la interactividad de la aplicación.

Dash proporciona clases HTML que nos permiten generar contenido HTML con Python. Para usar estas clases, necesitamos importar dash_core_components y dash_html_components.

* Usamos la clase Div de dash_html_components para crear un HTML Div. Luego usamos los componentes HTML para generar componentes HTML como H1, H2, etc. dash_html_components tiene todas las etiquetas HTML.
* Para crear un gráfico en nuestro diseño, usamos la clase Graph de dash_core_components. Graph representa visualizaciones de datos interactivas usando plotly.js. La clase Graph espera un objeto de figura con los datos que se trazarán y los detalles del diseño.

## **Creacion de Base de Proyecto**

Cree un proyecto llamado MyDashAPP y añada un archivo app.py donde se escribirá el código principal de la aplicación. En este caso, podemos emplear VSCODE o el IDE PyCharm.

Agregue las siguientes líneas de código en el archivo app.py,





```
1 from dash import Dash, html, dcc
2 import plotly.express as px
3 import pandas as pd
4 app = Dash(__name__)
5 app.layout = html.Div(" Hola Dash! :D ")
6 if __name__ == '__main__':
7   app.run(debug=True)
```
