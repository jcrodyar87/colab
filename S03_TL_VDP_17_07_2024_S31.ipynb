## **1.4 KDEPLOT**

El Estimador de Densidad del Kernel (KDE) es una tcnica estadstica que se utiliza para suavizar un conjunto de datos y obtener una visin ms precisa de su distribucin. Aunque se utiliza con frecuencia en diversos campos como la economa, la biologa y la informtica, su aplicacin en la ciencia del deporte es particularmente intrigante. Este artculo tiene como objetivo explicar qu es el KDE, cmo se aplica y por qu puede ser una herramienta valiosa para los cientficos del deporte, especialmente en el entrenamiento de atletas de resistencia.

**QUE ES EL KDE?**

KDE es una forma de estimar la funcin de densidad de probabilidad (PDF) de una variable aleatoria continua. A diferencia de los histogramas, que son rudimentarios y dependen del tamao del intervalo, el KDE proporciona una forma suavizada y ms flexible para entender la distribucin de los datos.

## **COMO FUNCIONA EL KDE?**

Para aplicar el KDE, siga los siguientes pasos:

1. Elija un Kernel: Funciones como el kernel Gaussiano y el kernel Epanechnikov son populares.
2. Seleccione el Ancho de Banda: Este parmetro controla el nivel de suavizado.
3. Calcule la Estimacin: Utilice la frmula del KDE para estimar la densidad en diferentes puntos.
4. Visualice los resultados: Represente grficamente la estimacin de la densidad para una interpretacin ms fcil.

kdeplots son Grficos de Estimacin de Densidad del Ncleo.

```python
# Variable 'total bill'
sns.kdeplot(tips['total_bill']) #plot kde
```

```python
#Variable 'tip'
sns.kdeplot(tips['tip'])
```

### **1.5.1 BARPLOT**

Es un grfico general que le permite agregar los datos categricos basados en alguna funcin, el valor predeterminado es la media:

### **1.5.3 BOXPLOT**

Los boxplots (diagrama de caja) y violin plots se utilizan para mostrar la distribucin de datos categricos. Un diagrama de caja (boxplots o grfico de caja y bigotes) muestra la distribucin de datos cuantitativos de una manera que facilita las comparaciones entre variables o entre niveles de una variable categrica. El cuadro muestra los cuartiles del conjunto de datos, mientras que los bigotes se extienden para mostrar el resto de la distribucin, a excepcin de los puntos que se determinan como valores atpicos utilizando un mtodo que es una funcin del rango intercuartlico.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow');
```

```python
# se pueden graficar de forma horizontal
sns.boxplot(data=tips,palette='rainbow',orient='h');
```

```python
# cambiar el color y ver varias variables (hue)
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm");
```

### **1.5.4 VIOLINPLOT**

Un plot de violn juega un papel similar a un box and whisker plot (diagrama de cajas y bigotes). Muestra la distribucin de datos cuantitativos a travs de varios niveles de una (o ms) variables categricas de modo que esas distribuciones se puedan comparar. A diferencia de un diagrama de caja, en el que todos los componentes de la grfica corresponden a los puntos de datos reales, la grfica del violn presenta una estimacin de la densidad del ncleo de la distribucin subyacente.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow');
```

```python
# Varias Variables
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1');
```

```python
# Varias variables
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1');
```

### **1.6.0  GRIDS**

Las grids son tipos generales de plots que le permiten mapear tipos de plots en filas y columnas de una cuadrcula, esto le ayuda a crear plots similares separadas por caractersticas.

```python
# Importar librerias
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris') #Importar el dataset
iris.head() #Ver los primeros 5 elementos de la tabla
```

### **1.6.1 PAIRGRID**

Pairgrid es un subplot grid para graficar relaciones por pares en un conjunto de datos.



```python
# solo el Grid
sns.PairGrid(iris);
```

```python
# Ahora se mapea el grid
g = sns.PairGrid(iris)
g.map(plt.scatter);
```

```python
# Mapear a arriba, abajo y diagonal
g = sns.PairGrid(iris) # crear una cuadricula
g.map_diag(plt.hist) #Histogramas en la diagonal
g.map_upper(plt.scatter) # Scatter plots en la parte superior
g.map_lower(sns.kdeplot); # Plots de densidad kde en la parte inferior
```

### **1.6.2 PAIRPLOT**

pairplot es una versin ms simple de PairGrid (se usa con bastante frecuencia)

```python
# La diagonal es un histograma
# las otras graficas son scatter plots
sns.pairplot(iris);
```

```python
# la diagonal con kde de los datos categorico
sns.pairplot(iris,hue='species',palette='rainbow');
```

## **USANDO MARCADORES**

Los argumentos kwargs lmplot pasan a **regplot** que es una forma mas general de lmplot(). regplot tiene un parametro scatter_kwds que se pasa a plt.scatter y puede modifcar los parametros

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm', markers=['o','v'], scatter_kws={'s':100})
```

## **USANDO GRID**

Podemos agregar una separacion mas variable a travs de columnas y filas con el uso de un grid. Indicando los argumentos col o row

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',row='sex',col='time',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```
