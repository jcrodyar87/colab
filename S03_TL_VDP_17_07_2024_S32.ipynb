# **Aplicación del Estimador de Densidad del Kernel en Ciencias del Deporte**

# **Introducción**

El Estimador de Densidad del Kernel (KDE) es una técnica estadística que se utiliza para suavizar un conjunto de datos y obtener una visión más precisa de su distribución. Aunque se utiliza con frecuencia en diversos campos como la economía, la biología y la informática, su aplicación en la ciencia del deporte es particularmente intrigante. Este artículo tiene como objetivo explicar qué es el KDE, cómo se aplica y por qué puede ser una herramienta valiosa para los científicos del deporte, especialmente en el entrenamiento de atletas de resistencia.



# **¿Qué es el KDE?**

KDE es una forma de estimar la función de densidad de probabilidad (PDF) de una variable aleatoria continua. A diferencia de los histogramas, que son rudimentarios y dependen del tamaño del intervalo, el KDE proporciona una forma suavizada y más flexible para entender la distribución de los datos.

# **¿Cómo funciona el KDE?**


Para aplicar el KDE, siga los siguientes pasos:

1. Elija un Kernel: Funciones como el kernel Gaussiano y el kernel Epanechnikov son populares.
2. Seleccione el Ancho de Banda: Este parámetro controla el nivel de suavizado.
3. Calcule la Estimación: Utilice la fórmula del KDE para estimar la densidad en diferentes puntos.
4. Visualice los resultados: Represente gráficamente la estimación de la densidad para una interpretación más fácil.


## **FUNCION DE DENSIDAD DE PROBABILIDAD**

La función de densidad de probabilidad de una variable aleatoria continua es una fórmula que ayuda a calcular la probabilidad de que una variable aleatoria continua tenga un valor que esté dentro de un intervalo específico.

## **Variable Aleatoria Continua**

Una variable aleatoria es una función que asigna un valor numérico, al resultado de un experimento aleatorio. Una variable aleatoria puede ser discreta o continua. Las variables aleatorias discretas son aquellas que presentan un número contable de valores; por ejemplo, el número de personas que viven en una casa (pueden ser 3, 5 o 9). Las variables aleatorias continuas son aquellas que presentan un número incontable de valores; por ejemplo, el peso de las vacas en una granja (una vaca puede pesar 632,12 kg, otra puede pesar 583,12312 kg, otra 253,12012 kg, otra 198,0876 kg y nunca terminaríamos de enumerar todos los posibles valores).

Una variable aleatoria continua, es aquella que puede asumir un número incontable de valores.  Por ejemplo, si realizamos el experimento de ir a una granja y estudiamos las características de las vaquitas, podemos definir la variable aleatoria B = peso de una vaca en la granja  (en kilogramos).

Alguna vaquita puede pesar 425,1872 kg; otra puede pesar 612,5874541 kg; otra puede pesar 545,897512121 kg. Si tomamos más vacas, podríamos tener más valores y nunca terminaríamos. Se conoce que el becerro más pequeño tiene un peso de 30 kg, y la vaca más grande tiene un peso de 1000 kg.

```python
import random
import numpy
import  statistics
```

```python
tiempos_atleta1= numpy.array([random.randrange(70, 80) for i in range(100)])
```

```python
tiempos_atleta2= numpy.array([random.randrange(77, 90) for i in range(100)])
```

## ATLETA 1 DISTRIBUCION NORMAL DE TIEMPOS - 100 MUESTRAS , MEDIA 75 SEG DESVIACION ESTANDAS 3

```python
tiempos_atleta1=numpy.array(statistics.NormalDist(75, 3).samples(100, seed=3652260728))
```

ATLETA 2 DISTRIBUCION NORMAL DE TIEMPOS- 100 MUESTRAS, MEDIA 81 SEG DESVIACION ESTANDAR DE 2

```python
tiempos_atleta2=numpy.array(statistics.NormalDist(81, 2).samples(100, seed=3652260728))
```

```python
tiempos_atleta1
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
```

```python
plt.plot(tiempos_atleta1)
```

```python
plt.plot(tiempos_atleta2)
```

```python
# Preparar los datos para KDE
tiempos = np.concatenate([tiempos_atleta1, tiempos_atleta2])
```

```python
tiempos.shape
```

```python
tiempos = tiempos[:, np.newaxis]
```

```python
tiempos.shape
```

```python
# Crear el objeto para el KDE y ajustarlo a los datos
kde = KernelDensity(kernel='gaussian', bandwidth=1.0).fit(tiempos)
```

```python
# Evaluar en un rango de tiempos mas rapido y mas lento
tiempos_evaluar = np.linspace(70, 90, 1000)[:, np.newaxis]
densidades_log = kde.score_samples(tiempos_evaluar)
densidades = np.exp(densidades_log)
```

```python
# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.hist(tiempos_atleta1, alpha=0.5, label='Atleta 1', bins=10, density=True)
plt.hist(tiempos_atleta2, alpha=0.5, label='Atleta 2', bins=10, density=True)
plt.plot(tiempos_evaluar, densidades, label='KDE')
plt.xlabel('Tiempo de carrera (minutos)')
plt.ylabel('Densidad')
plt.title('Distribución de tiempos de carrera para atletas de resistencia')
plt.legend()
plt.show()
```

## **Ancho de banda óptimo Por qué es crítico?**

1. Subajuste (Underfitting): Un ancho de banda demasiado grande hará que la estimación sea demasiado suave, lo que podría ocultar detalles importantes en la distribución de los datos. Esto podría llevar a interpretaciones erróneas y a decisiones basadas en una visión simplificada de los datos.
2. Sobreajuste (Overfitting): Por otro lado, un ancho de banda demasiado pequeño hará que la estimación siga el ruido en los datos, lo que también podría llevar a interpretaciones erróneas. En este caso, se capturan “picos” y “valles” que pueden no ser representativos de la distribución subyacente, sino más bien del ruido en los datos.

## **¿Cómo elegir el ancho de banda óptimo?**

Existen varias estrategias para elegir un ancho de banda óptimo:

1. Métodos de validación cruzada: Se pueden usar técnicas como la validación cruzada para encontrar el ancho de banda que minimiza un determinado criterio de error.
2. Reglas empíricas: Algunas reglas como la “regla de Silverman” ofrecen una estimación rápida pero menos precisa del ancho de banda óptimo basada en estadísticas descriptivas de los datos.
3. Experiencia de dominio: En campos aplicados como la ciencia del deporte, la experiencia y el conocimiento del dominio pueden guiar la elección del ancho de banda para capturar aspectos relevantes de la distribución de los datos.


Dado que cada conjunto de datos y cada pregunta de investigación son únicos, la elección del ancho de banda suele requerir un equilibrio entre la precisión estadística y la interpretabilidad práctica. Es por eso que la elección del ancho de banda es una consideración crítica en la aplicación del KDE.

```python
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV
tiempos_atleta=tiempos
# Parámetros del Grid Search
parametros = {'bandwidth': np.linspace(0.1, 2.0, 30)}
grid = GridSearchCV(KernelDensity(kernel='gaussian'), parametros, cv=5)
grid.fit(tiempos_atleta)

# Mejor ancho de banda
mejor_bandwidth = grid.best_params_['bandwidth']
print(f"El mejor ancho de banda encontrado es: {mejor_bandwidth}")

# Aplicar KDE con el mejor ancho de banda
kde = KernelDensity(bandwidth=mejor_bandwidth, kernel='gaussian')
kde.fit(tiempos_atleta)

# Evaluar en un rango de tiempos
tiempos_evaluar = np.linspace(70, 90, 1000)[:, np.newaxis]
log_densidad = kde.score_samples(tiempos_evaluar)
densidad = np.exp(log_densidad)

# Visualizar
plt.figure(figsize=(10, 6))
plt.hist(tiempos_atleta, bins=10, density=True, alpha=0.5, label='Histograma de tiempos')
plt.plot(tiempos_evaluar, densidad, label='KDE con ancho de banda óptimo')
plt.xlabel('Tiempo de carrera (minutos)')
plt.ylabel('Densidad')
plt.title('Estimación de KDE con ancho de banda óptimo')
plt.legend()
plt.show()
```

```python
# Preparar los datos para KDE
tiempos = np.concatenate([tiempos_atleta1, tiempos_atleta2])
tiempos = tiempos[:, np.newaxis]
# Crear el objeto para el KDE y ajustarlo a los datos
kde = KernelDensity(kernel='gaussian', bandwidth= 1.6724137931034482).fit(tiempos)
# Evaluar en un rango de tiempos mas rapido y mas lento
tiempos_evaluar = np.linspace(70, 90, 1000)[:, np.newaxis]
densidades_log = kde.score_samples(tiempos_evaluar)
densidades = np.exp(densidades_log)
# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.hist(tiempos_atleta1, alpha=0.5, label='Atleta 1', bins=10, density=True)
plt.hist(tiempos_atleta2, alpha=0.5, label='Atleta 2', bins=10, density=True)
plt.plot(tiempos_evaluar, densidades, label='KDE')
plt.xlabel('Tiempo de carrera (minutos)')
plt.ylabel('Densidad')
plt.title('Distribución de tiempos de carrera para atletas de resistencia')
plt.legend()
plt.show()
```
